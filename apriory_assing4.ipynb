```python
#viewing the dataset 
import pandas as pd
#import matplotlib.pyplot as plt


spreadsheet = pd.read_csv('/content/Cosmetics - Cosmetics.csv')

spreadsheet
```

```python
#q1  frequnet itemset 
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
tr = TransactionEncoder()
tr_arr = tr.fit(spreadsheet).transform(spreadsheet)
df = pd.DataFrame(tr_arr, columns=tr.columns_)
df
```

```python
df = df.dropna()
```

```python
from mlxtend.frequent_patterns import apriori
frequent_itemsets = apriori(df, min_support = 0.3, use_colnames = True)
frequent_itemsets
```

```python

#viewing the dataset 
import pandas as pd
#import matplotlib.pyplot as plt

spreadsheet = pd.read_excel('/content/Cosmetics.xlsx')

spreadsheet
```

```python
spreadsheet=spreadsheet.replace({'Yes':True ,'No' :False })
spreadsheet
```

```python




from mlxtend.frequent_patterns import apriori, association_rules
spreadsheet
frq_set=apriori(spreadsheet, min_support=0.3,use_colnames=True)
frq_set
```

```python
rules = association_rules(frq_set, metric ="confidence", min_threshold = 0.7) 
rules = rules.sort_values(['confidence'], ascending =[ False]) 
rules
```
